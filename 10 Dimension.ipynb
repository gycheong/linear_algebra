## Review from last notes

**Proposition**. Given vectors $\boldsymbol{v}_1, \dots, \boldsymbol{v}_m \in \mathbb{F}^n$, let $A = [\boldsymbol{v}_1 | \cdots | \boldsymbol{v}_m]$. Then
1. $\boldsymbol{v}_1, \dots, \boldsymbol{v}_m$ are linearly independent if and only if $\mathrm{REF}(A)$ has a pivot $1$ in every column;
2. $\boldsymbol{v}_1, \dots, \boldsymbol{v}_m$ span $\mathbb{F}^n$ if and only if $\mathrm{REF}(A)$ has a pivot $1$ in every row;
3. $\boldsymbol{v}_1, \dots, \boldsymbol{v}_m$ form a basis of $\mathbb{F}^n$ if and only if $\mathrm{REF}(A)$ has a pivot $1$ in every column and every row.

**Remark**. Note that the situation in Case 3 above forces $m = n$ and $\mathrm{REF}(A) = I_n$. From Case 1, it also follows that:

**Proposition**. 
1. A set of linearly independent vectors in $\mathbb{F}^n$ has size at most $n$. 
2. Any spanning set of $\mathbb{F}^n$ has at least $n$ vectors.
3. A basis of $\mathbb{F}^n$ has exactly $n$ vectors.

## Dimension

**Theorem**. Let $V$ be a vector space over $\mathbb{F}$. If $V$ is spanned by finitely many vectors in $V$, then for any two bases $\mathcal{B}$ and $\mathcal{B}'$, we have $|\mathcal{B}| = |\mathcal{B}'|$.

**Remark/Definition**. Note that we already know that once we have a finite spanning set of $V$ then we can extract a basis from it. The above theorem says that although we can have different bases to $V$, their sizes must be the same; such a size is called the **dimension** of $V$, and we write $\dim(V)$ to mean it.

In fact, even when $V$ is not necessarily generated by finitely many vectors, it turns out that $V$ does have a basis and any two bases have the same cardinality. However, we will not discuss much about such vector spaces in this class. Such vector spaces are said to be "infinite-dimensional", and they arise quite commonly in more advanced real/complex analysis.

The proof of Theorem provides a chance to study the following important lemma:

**Lemma**. Let $V$ and $W$ be finite dimensional vector spaces over $\mathbb{F}$. Say $\boldsymbol{v}_1, \dots, \boldsymbol{v}_n$ form a basis for $V$. Fix any vectors $\boldsymbol{w}_1, \dots, \boldsymbol{w}_n \in W$. Then

1. we have a unique linear map $\phi : V \rightarrow W$ such that $\phi(\boldsymbol{v}_i) = \boldsymbol{w}_i$ for $1 \leq i \leq n$;
2. if $\boldsymbol{w}_1, \dots, \boldsymbol{w}_n$ form a basis for $W$, then $\phi$ is an isomorphism.

*Proof*. Note that in order for $\phi$ to satsify $\phi(\boldsymbol{v}_i) = \boldsymbol{w}_i$ for $1 \leq i \leq n$, we have no choice but to define:

$$\phi(a_1\boldsymbol{v}_1 + \cdots + a_n\boldsymbol{v}_n) := a_1\boldsymbol{w}_1 + \cdots + a_n\boldsymbol{w}_n$$ 

for any $a_1, \dots, a_n \in \mathbb{F}$. Then one can easily check such $\phi : V \rightarrow W$ is linear, which shows the first claim.

For the second claim, if $\boldsymbol{w}_1, \dots, \boldsymbol{w}_n$ form a basis for $W$, then using the first claim, we can construct a unique linear map $\psi : W \rightarrow V$ such that $\psi(\boldsymbol{w}_i) = \boldsymbol{v}_i$ for $1 \leq i \leq n$. One can check that $\psi \circ \phi = \mathrm{id}_V$ and $\phi \circ \psi = \mathrm{id}_W$, showing that $\psi$ is the inverse map of $\phi$. $\Box$

*Proof of Theorem*. Let $n = |\mathcal{B}|$ and $m = |\mathcal{B}'|$ and write
* $\mathcal{B} = \{\boldsymbol{v}_1, \dots, \boldsymbol{v}_n\}$ and
* $\mathcal{B}' = \{\boldsymbol{w}_1, \dots, \boldsymbol{w}_m\}$

Consider the linear map $\phi : \mathbb{F}^n \rightarrow V$ given by $\phi(\boldsymbol{e}_j) = \boldsymbol{v}_j$ for $1 \leq j \leq n$. This is an isomorphism by Lemma. Since $\phi^{-1} : V \rightarrow \mathbb{F}^n$ is also an isomorphism, it follows that $\phi^{-1}(\boldsymbol{w}_1), \dots, \phi^{-1}(w_m)$ form a basis for $\mathbb{F}^n$. Hence, it follows that $m = n$. $\Box$


**Exercise** Let $V$ be a finite-dimensional vector space and $n := \dim(V)$. Show that

1. a set of linearly independent vectors in $V$ has size at most $n$;

**Theorem (basis extension theorem)**. Let $V$ be a finite-dimensional vector space, and $\boldsymbol{v}_1, \dots, \boldsymbol{v}_r$ some linearly independent vectors in $V$. Then there are $\boldsymbol{v}_{r+1}, \dots, \boldsymbol{v}_n \in V$ such that $\boldsymbol{v}_1, \dots, \boldsymbol{v}_r, \boldsymbol{v}_{r+1}, \dots, \boldsymbol{v}_n$ form a basis for $V$.

*Proof*. We pick $\boldsymbol{v}_{r+1}$ that is NOT in the span of $\boldsymbol{v}_1, \dots, \boldsymbol{v}_r$. (If we cannot, then $r = n$, so we would be done.) We know from way back that the extended set $\{\boldsymbol{v}_1, \dots, \boldsymbol{v}_r, \boldsymbol{v}_{r+1}\}$ is linearly independent, so we can continue this process so that we get to extend to the $n$ linearly independent vectors including the given linearly independent vectors. (Can you see why $n$ linearly independent vectors span the whole space $\mathbb{F}$? Hint: Use Lemma above.) $\Box$
